```python
# Load the training dataset
train_data = pd.read_csv('twitter_training.csv')

# Rename the columns for clarity
train_data.columns = ['ID', 'Topic', 'Sentiment', 'Text']

# Clean the data by dropping missing values
train_data_clean = train_data.dropna()

# Display the first few rows of the cleaned data
train_data_clean.head()
```

```python
# Check the distribution of sentiments
sentiment_counts = train_data_clean['Sentiment'].value_counts()

# Plotting the sentiment distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='viridis')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# If you want to analyze specific topics, filter the data
specific_topic = 'Borderlands'  # Example topic
topic_data = train_data_clean[train_data_clean['Topic'] == specific_topic]

# Check sentiment distribution for a specific topic
topic_sentiment_counts = topic_data['Sentiment'].value_counts()

# Plotting the sentiment distribution for the specific topic
plt.figure(figsize=(8, 6))
sns.barplot(x=topic_sentiment_counts.index, y=topic_sentiment_counts.values, palette='coolwarm')
plt.title(f'Sentiment Distribution for {specific_topic}')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
