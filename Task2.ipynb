```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
titanic_data = pd.read_csv('/content/train.csv')

# Data Cleaning
# Checking for missing values
missing_values = titanic_data.isnull().sum()

# Dropping columns with too many missing values or that are not useful for analysis
titanic_data = titanic_data.drop(['Cabin', 'Ticket'], axis=1)

# Filling missing values in 'Age' with the median
titanic_data['Age'] = titanic_data['Age'].fillna(titanic_data['Age'].median())

# Filling missing values in 'Embarked' with the mode
titanic_data['Embarked'] = titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0])

# Checking the cleaned data
print(titanic_data.info())

# Exploratory Data Analysis (EDA)

# 1. Distribution of Survival
plt.figure(figsize=(8,6))
sns.countplot(x='Survived', data=titanic_data)
plt.title('Distribution of Survival')
plt.show()

# 2. Survival Rate by Sex
plt.figure(figsize=(8,6))
sns.countplot(x='Survived', hue='Sex', data=titanic_data)
plt.title('Survival Rate by Sex')
plt.show()

# 3. Survival Rate by Passenger Class
plt.figure(figsize=(8,6))
sns.countplot(x='Survived', hue='Pclass', data=titanic_data)
plt.title('Survival Rate by Passenger Class')
plt.show()

# 4. Age Distribution
plt.figure(figsize=(8,6))
sns.histplot(titanic_data['Age'], kde=True)
plt.title('Age Distribution of Passengers')
plt.show()

# 5. Survival Rate by Age
plt.figure(figsize=(8,6))
sns.histplot(titanic_data[titanic_data['Survived'] == 1]['Age'], color='green', kde=True, label='Survived')
sns.histplot(titanic_data[titanic_data['Survived'] == 0]['Age'], color='red', kde=True, label='Not Survived')
plt.legend()
plt.title('Survival Rate by Age')
plt.show()

# Identifying patterns or trends:
# The code above will help in identifying relationships such as:
# - The impact of gender, passenger class, and age on survival rates
# - General trends in age distribution and survival
```
